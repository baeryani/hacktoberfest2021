```python
import billboard
#import lyricsgenius
import datetime
import pandas as pd
#genius = lyricsgenius.Genius('J2GYbVYl2VoFWPkmhY6wjkJUDhMnpth-NjGxuUYx-9Y8sbMSoRiJB7LwvnVHzSfj')
```

```python
year = []
songName = []
artistName = []
```

```python
start_date = datetime.date(1970, 1, 1)
end_date = datetime.date(1989, 12, 31)
delta = datetime.timedelta(days=7)
dates = []
while start_date <= end_date:
    dates.append(str(start_date))
    start_date += delta
```

```python
dates
```

```python
#Genius can handle only 200 requests at a time, so make sure to add the requests in small batches
```

```python
#billboard charts are updated weekly to we will the dates in that format from 2000-2010 from the above created list
for date in dates:
    print("------------------------------Starting for date: ", date,"------------------------------")
    chart = billboard.ChartData('hot-100', date=date)
    for song in chart:
        if song not in songName:
            year.append(date[:4])
            s = str(song.title)
            songName.append(song.title)
            artistName.append(song.artist)
```

```python
dataFrame = {
    "Year": year,
    "Song": songName,
    "Artist/Band": artistName
}
```

```python
dataFrame
```

```python
data = pd.DataFrame.from_dict(dataFrame)
```
